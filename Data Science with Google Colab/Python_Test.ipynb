```python
max = clean_df['Mid-Career Median Salary'].max()
index = clean_df['Mid-Career Median Salary'].idxmax()
college = clean_df['Undergraduate Major'].loc[index]
print(f'The College that has the highest salary in mid carrer is {college} with {round(max)}$')
```

```python
low = clean_df['Starting Median Salary'].min()
index_low = clean_df['Starting Median Salary'].idxmin()
college = clean_df['Undergraduate Major'][index_low]
print(f'The College that has the lowest salary is {college} with {round(low)}$')
```

```python
low = clean_df['Mid-Career Median Salary'].min()
index_low = clean_df['Mid-Career Median Salary'].idxmin()
college = clean_df['Undergraduate Major'].loc[index_low]
print(f'The College that has the lowest salary is {college} with {round(low)}$')
```

```python
clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
high_risk = clean_df.sort_values('Mid-Career 90th Percentile Salary',ascending=False)
high_risk[['Undergraduate Major','Mid-Career 90th Percentile Salary']].head()
```

```python
high_risk = clean_df.sort_values('Spread',ascending=False)
high_risk[['Undergraduate Major','Spread']].head()
```
